```python
image_dirctory = '/content/archive/datasets/'
```

```python
no_tumor_image = os.listdir(image_dirctory + 'no/')
yes_tumor_image = os.listdir(image_dirctory + 'yes/')
```

```python
import numpy as np
dataset = []
label = []
for i, image_name in enumerate(no_tumor_image):
  if(image_name.split('.')[1] == 'jpg'):
    image = cv2.imread(image_dirctory+'no/'+image_name)
    image = Image.fromarray(image,'RGB')
    image = image.resize((64,64))
    dataset.append(np.array(image))
    label.append(0)
```

```python
for i, image_name in enumerate(yes_tumor_image):
  if(image_name.split('.')[1] == 'jpg'):
    image = cv2.imread(image_dirctory+'yes/'+image_name)
    image = Image.fromarray(image,'RGB')
    image = image.resize((64,64))
    dataset.append(np.array(image))
    label.append(1)
```

```python
print(len(dataset))
print(len(label))
```

```python
dataset = np.array(dataset)
label = np.array(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_trian, y_test = train_test_split(dataset,label,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
```

```python
import tensorflow as tf
from tensorflow import keras 
from keras.utils.np_utils import normalize
```

```python
print(X_train.shape)
```

```python
X_train = normalize(X_train, axis=1)
X_test = normalize(X_test,axis=1)

y_trian = to_categorical(y_trian, num_classes=2)
y_test = to_categorical(y_test, num_classes=2)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D,MaxPool2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.utils.np_utils import to_categorical
```

```python

#cnn model
model=Sequential()

model.add(Conv2D(32, (3,3), input_shape=(64, 64, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3), kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Conv2D(64, (3,3), kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('softmax'))
'''
[[1,2,3,4],
[4,5,6,7],
[8,9,10,11],
[8,9,10,11]]
'''
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.fit(X_train, y_trian, 
batch_size=128, 
verbose=1, epochs=15, 
validation_data=(X_test, y_test),
shuffle=False)
```

```python
model.save('BrainTumorbest_model1.h5')
```

```python
from keras.models import load_model
model=load_model('/content/BrainTumorbest_model1.h5')
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('\n', 'Test accuracy:', score[1]*100,"%")
```
